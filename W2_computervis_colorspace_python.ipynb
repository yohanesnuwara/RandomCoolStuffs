```
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```
nemo = cv2.imread('/content/nemo.jpg')
cv2_imshow(nemo)
plt.show()
```

```
bgr_nemo = cv2.cvtColor(nemo, cv2.COLOR_RGB2BGR)
cv2_imshow(bgr_nemo)
plt.show()
```

```
hsv_nemo = cv2.cvtColor(nemo, cv2.COLOR_RGB2HSV)
cv2_imshow(hsv_nemo)
plt.show()
```

```
gray_img = cv2.cvtColor(nemo, cv2.COLOR_RGB2GRAY)
cv2_imshow(gray_img)
plt.show()
```

```
yuv_img = cv2.cvtColor(nemo, cv2.COLOR_BGR2YUV)
cv2_imshow(yuv_img)
plt.show()
```

```
# splitting into r, g, b channels

b = cv2_imshow(nemo[:, :, 0])
g = cv2_imshow(nemo[:, :, 1])
r = cv2_imshow(nemo[:, :, 2])
```

```
# merge the channels

b, g, r = cv2.split(nemo)
merge_rgb_nemo = cv2.merge([r,g,b])
cv2_imshow(merge_rgb_nemo)
```

## Task 2. Separating RGB channels

Image: Prokudin-Gorskii photograph

![image](https://user-images.githubusercontent.com/51282928/81806382-951a4e80-9546-11ea-8b12-e6532b703196.png)

```
landscape = cv2.imread('/content/image.jpg')
cv2_imshow(landscape)
```

```
# get the size of the picture
height, width, channels = landscape.shape
print('Size, height:', height, 'and width:', width)

# divide into 3 equal parts (b, g, r)
rr = height / 3

b = landscape[1:int(rr), 0:int(width)]
g = landscape[int(rr)+1:2*int(rr), 0:int(width)]
r = landscape[2*int(rr)+1:3*int(rr), 0:int(width)]

# the shape of r, g, and b must be the same

plt.figure(figsize=(17,17))

plt.subplot(1,3,1)
plt.imshow(b)
plt.title('B channel')
plt.xticks([]); plt.yticks([])

plt.subplot(1,3,2)
plt.imshow(g)
plt.title('G channel')
plt.xticks([]); plt.yticks([])

plt.subplot(1,3,3)
plt.imshow(r)
plt.title('R channel')
plt.xticks([]); plt.yticks([])

plt.show()
```

```
# b = cv2_imshow(b[:, :, 0])
# g = cv2_imshow(g[:, :, 1])
# r = cv2_imshow(r[:, :, 2])

B, zero, zero = cv2.split(b)
zero, G, zero = cv2.split(g)
zero, zero, R = cv2.split(r)

landscape_merge = cv2.merge([B,G,R])
cv2_imshow(landscape_merge)
```
